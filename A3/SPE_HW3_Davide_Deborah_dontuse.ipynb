```python
import numpy as np
import queue

class MM1Queue:
    """
    An M/M/1 queue with a Poisson arrival process and an exponential service time distribution.
    """

    def __init__(self, lambda_, mu):
        """
        Initialize the M/M/1 queue with the arrival rate lambda_ and the service rate mu.
        """
        self.lambda_ = lambda_
        self.mu = mu
        self.queue = queue.Queue(maxsize=1)
        self.arrival_time = 0
        self.departure_time = 0
        self.num_arrivals = 0
        self.num_departures = 0
        self.system_time = 0

    def generate_arrival(self):
        """
        Generate a new arrival event with an exponential interarrival time.
        """
        interarrival_time = np.random.exponential(scale=1/self.lambda_)
        self.arrival_time = self.system_time + interarrival_time
        self.num_arrivals += 1

    def generate_departure(self):
        """
        Generate a new departure event with an exponential service time.
        """
        service_time = np.random.exponential(scale=1/self.mu)
        self.departure_time = self.system_time + service_time
        self.num_departures += 1

    def simulate(self, max_time):
        """
        Simulate the M/M/1 queue for a maximum time of max_time.
        """
        while self.system_time < max_time:
            # Generate a new arrival event
            self.generate_arrival()

            # Check if the queue is empty
            if self.queue.empty():
                # If the queue is empty, schedule the new arrival
                self.system_time = self.arrival_time
            else:
                # If the queue is not empty, check if the new arrival can be served immediately
                if self.arrival_time < self.departure_time:
                    # If the new arrival can be served immediately, schedule the departure
                    self.system_time = self.arrival_time
                    self.queue.get()
                    self.generate_departure()
                else:
                    # If the new arrival cannot be served immediately, add it to the queue
                    self.queue.put(self.arrival_time)

            # Check if there is a departure event
            if self.departure_time <= self.system_time:
                # If there is a departure event, remove the served customer from the queue
                self.queue.get()
                self.generate_departure()

            # Update the system time
            self.system_time += 1

# Example usage
lambda_ = 1.0  # Arrival rate
mu = 2.0  # Service rate
max_time = 1.0  # Maximum simulation time
queue = MM1Queue(lambda_, mu)
queue.simulate(max_time)
print(f"The number of arrivals is {queue.num_arrivals}")
print(f"The number of departures is {queue.num_departures}")
```
